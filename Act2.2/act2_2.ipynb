```python
df = pd.read_csv("Athenas_Grecia_Dt.csv")
df.head()
```

```python
df.info() 
```

```python
#Registros del host que el response rate sea mayor a 95% y que no sean superhosts
filtro9=df[(df["host_response_rate"] > "95%") & (df["host_is_superhost"] == "f")]

filtro9.to_csv("Filtro9.csv")
filtro9
```

```python
#Los registros que no tengas 1 baño 
filtro10 = df[~(df["bathrooms"] == 1)]

filtro10.to_csv("Filtro10.csv")
filtro10 
```

```python
#Eliminar un signo de una columna 
df['price']=df['price'].str.replace('$', '')
df['price']=df['price'].str.replace(',', '')
```

```python
#Conversión de tipo de dato de columna de tipo Object a Float
df['price']= df['price'].astype(float)
df
```

```python
#Lors registros que este en un precio de 100 y 200
filtro11=df[(df["price"] <= 200) & (df["price"] >= 100)]


filtro11.to_csv("Filtro11.csv")
filtro11
```

```python
#Los registros que sean kisting count mayor a 10 y que tenga un review score menor que 4.5
#No tengo las variables de instant_bookable y la de host_identity_verified
filtro12=df[(df["host_listings_count"] < 10) & (df["review_scores_value"] > 4.5 )]

filtro12.to_csv("Filtro12.csv")
filtro12
```

```python
#Los registros de los tipo de cuarto hotel room y private room 
filtro13=df[(df["room_type"] == "Hotel room") | (df["room_type"] == "Private room")]

filtro13.to_csv("Filtro13.csv")
filtro13
```

```python
#Los registros que tengan más de 4 accomodates
filtro14=df[df["accommodates"] > 4]

filtro14.to_csv("Filtro14.csv")
filtro14
```
